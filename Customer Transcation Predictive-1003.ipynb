```python
#Training and Testing
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('target',axis=1)
y=df['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
# Drop the 'ID_code' column before scaling
x_train_numeric = x_train.drop('ID_code', axis=1)
x_test_numeric = x_test.drop('ID_code', axis=1)

sc.fit(x_train_numeric)
x_train_sc=sc.transform(x_train_numeric)
x_test_sc=sc.transform(x_test_numeric)
```

```python
#Dimesion Reduction Technique(PCA)
```

```python
from sklearn.decomposition import PCA
sc=StandardScaler()
pca = PCA(n_components=0.95)

x_train_pca = pca.fit_transform(x_train_sc)
x_test_pca = pca.transform(x_test_sc)

print("Original Features:", x_train.shape[1])
print("Reduced Features:", x_train_pca.shape[1])
```

```python
#Model Building
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train_sc,y_train)
```

```python
from sklearn.metrics import accuracy_score
lr_pred=lr.predict(x_test_sc)
accuracy_score=accuracy_score(y_test,lr_pred)
print(accuracy_score)
```

```python
lr_pred
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train_sc,y_train)
```

```python
from sklearn.metrics import accuracy_score
dt_pred=dt.predict(x_test_sc)

accuracy_score=accuracy_score(y_test,dt_pred)
print(accuracy_score)
```

```python
import xgboost as xgb
```

```python
xgb_model=xgb.XGBClassifier()
xgb_model.fit(x_train_sc,y_train)
```

```python
from sklearn.metrics import accuracy_score
xgb_pred=xgb_model.predict(x_test_sc)

accuracy_score=accuracy_score(y_test,xgb_pred)
print(accuracy_score)
```

```python
# Model Evaluation(Accuracy , Classification_report and Confusion matrix)
```
